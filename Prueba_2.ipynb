```python
# Definir diccionario de datos
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
# Definir índices
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

```python
# Borrar la fila 'k' y entregar nuevamente el dataframe original
df.drop(['k'], inplace=True)
```

```python
# Ordenar el dataframe df por la columna name en orden ascendente
df.sort_values(by='name', inplace=True)
```

1. Importarlo como nombre athletes

```python
# Importar dataset
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
```

2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.

```python
# Filtrar el dataframe por los atletas que hayan ganado por lo menos una medalla
# Opción 1:
# Generar columna 'total_medals' con la suma de los valores de las columnas 'bronze', 'silver' y 'gold'
athletes['total_medals'] = athletes.loc[:,['gold', 'silver', 'bronze']].sum(axis = 1)
# Filtrar el dataframe por los registros en los que total_medals sea mayor que 0
athletes.loc[athletes['total_medals'] > 0]
```

```python
# Mostrar DataFrame con los registros donde se tiene almenos una medalla
athletes.loc[(athletes["gold"] > 0) | (athletes["silver"] > 0) | (athletes["bronze"] > 0)]
```

3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

```python
#Agregar nueva columna de BMI al DF la cual se calcula como el peso dividido la estatura al cuadrado
athletes['BMI'] = athletes.weight / athletes.height**2
athletes
```

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

```python
#Quitar el id con la función drop()
athletes.drop(["id"], axis=1, inplace=True)
```

```python
#Agrupar por deporte con el cálculo de la media del resto de características
athletes_mean = athletes.groupby("sport").mean()
#Ordenar por peso de manera descendente
athletes_mean.sort_values(by="weight", ascending=False)
```

```python
#Usar idxmax() para mostrar el deporte con el mayor peso promedio de sus jugadores
sport_max = athletes_mean.weight.idxmax()
sport_max
```

5. ¿Cuál es el pais con más participación femenina?

```python
#Filtrar por sexo femenino
female_athletes = athletes.loc[athletes["sex"] == "female"]
```

```python
female_athletes
```

```python
#Contar el número de mujeres por nacionalidad
country_counts = female_athletes['nationality'].value_counts()
#Usar idxmax() para mostrar el país con el mayor número de mujeres participantes
country_max = country_counts.idxmax()
country_max
```

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Importar dataset
nba = pd.read_csv("https://raw.github.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
nba.head()
nba
```

```python
# Eliminar las filas con datos vacíos
nba = nba.dropna(how='any')
nba
```

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

```python
# Agrupar por la columna 'Team' y y sumar los datos en cada grupo
nba_salary = nba.groupby(by="Team").sum()
# Ordenar de manera descendente por la columna 'Salary'
nba_salary = nba_salary.sort_values(by="Salary", ascending=False)
# Mostrar el dataframe actualizado
nba_salary[["Salary"]]
```

```python
# Definr el equipo con el valor máximo en 'Salary'
team_max = nba_salary.Salary.idxmax()
team_max
print(f"El equipo con un salario más alto es {team_max}, con una nómina total de {nba_salary.Salary.max()}")
```

2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.

```python
# Agrupar en tres segmentaciones por el valor en 'Age'
nba['Age-Group'] = pd.cut(nba.Age, bins=3, labels=['inf', 'med', 'sup'])
# Mostrar el dataframe actualizado
nba
```

```python
# Mostrar los límites máximo y mínimo de cada grupo
nba.groupby("Age-Group").Age.agg(['min', 'max']).sort_values(by='min')
```

```python
# Generar un conteo de los datos incluidos en cada grupo
nba[["Age-Group"]].value_counts()
```

```python
# Generar un nuevo dataframe con la agrupación generada 'Age-Group' como índice y la columna 'Position' con un conteo de los datos que agrupan las dos variables
nba_age_position = nba.groupby(["Age-Group", "Position"]).Name.count().unstack()
# Mostrar el dataframe actualizado
nba_age_position
```

```python
# Generar una gráfica de barras apiladas
nba_age_position.plot(kind = "bar",
                      stacked = 'True');
```

3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Formatear display a dos cifras decimales, para evitar notación científica
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# Generar tabla pivot con las variables 'Team', 'Position' y el promedio de la variable 'Salary'
nba.pivot_table(index='Team',
                columns='Position',
                aggfunc={'Salary': 'mean'})
```
